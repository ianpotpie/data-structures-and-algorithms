# Linked Lists

A linked list is a linear data structure, in which the elements are not stored at contiguous memory locations. The
elements in a linked list are linked using pointers

```python
from SinglyLinkedList import SinglyLinkedList
from random import random

linked_list = SinglyLinkedList()
for i in range(5):
    r = random()
    linked_list.push(r)


print(linked_list.stringify())

graph = linked_list.graphify()
graph
```
